```python
import numpy
import scipy
import matplotlib.pyplot as plt
```

```python
#Simulation #1 Parameters
n = 20
n_sim = 1000
max_gen = 2000
```

```python
#Simulation #2 Parameters
n = 200
n_sim = 1000
max_gen = 2000
```

```python
#Simulation
simulations = []
times_to_fixation = []
for s in range(n_sim):
    p = 0.5
    black_alleles = []
    fixated = False
    for i in range(max_gen):
        num_black = numpy.random.binomial(n, p)
        black_alleles.append(num_black)
        p = num_black/n
        if (p == 1 or p == 0) & (fixated == False): 
            times_to_fixation.append(i)
            fixated = True
    simulations.append(black_alleles)
```

```python
plt.plot(numpy.array(simulations).T)
plt.show()
```

This is a Markov Process with absorbing boundaries.

```python
numpy.mean(times_to_fixation)
```

```python
def S(n_black, N):
    n_white = N - n_black
    p_black = (n_black)/(n_black + n_white)
    p_white = (n_white)/(n_black + n_white)
    return -(p_black*numpy.log(p_black) + p_white*numpy.log(p_white))

def T(n_alleles, n_population, n_black):
    return n_alleles*n_population*S(n_black, n_population)
```

```python
# Calculating time to fixation for Simulation #1
T(n_alleles = 2, n_population = 20, n_black = 10)
```

```python
# Calculating time to fixation for Simulation #2
T(n_alleles = 2, n_population = 200, n_black = 100)
```
